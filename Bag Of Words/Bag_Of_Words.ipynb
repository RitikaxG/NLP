######  Bag Of Words: Count Vectorizer
- Sparse Representation, that may consume too much memory & computing resources
- Doesn't capture meaning of words properly.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam.csv")
df.head()
```

```python
df.Category.value_counts()
```

```python
#On category column apply some transformation and generate new column
df['spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam , test_size = 0.2) 
```

```python
#Bag of word model for numeric represntation
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()

X_train_cv = v.fit_transform(X_train.values)
X_train_cv #gives sparse matrix of type numpy.int64
```

```python
#Building an ML Model
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
classification_report(y_test,y_pred)
```

```python
emails = [
    'Hey mohan, can we get together to watch football game tomorrow?',
    "Upto 20% discount on parking, exclusive offer just for you. Don't miss this reward!"
]

emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
#Alternative Method
#Pipeline is the convenient API
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
classification_report(y_test,y_pred)
```
